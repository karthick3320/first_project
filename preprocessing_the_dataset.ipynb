## droping unwanted features 

```python
dataset.columns.values

columns_to_remove = ['people_ID','Designation','Name','Insurance','salary']

dataset_filtered=dataset.drop(columns_to_remove,axis=1)
```

## missing values checkpoint _1

```python
copied_dataset = dataset_filtered.copy()
```

```python
copied_dataset['comorbidity'] = copied_dataset['comorbidity'].map({"Hypertension":"Disease_vulnerable","Diabetes":"Disease_vulnerable","Coronary Heart Disease":"Disease_vulnerable","None":"NO","nan":"NO"})
```

```python
copied_dataset['comorbidity'] = copied_dataset['comorbidity'].fillna(value = "NO")
```

```python
copied_dataset['cardiological pressure'] = copied_dataset['cardiological pressure'].fillna(value = 'Normal')
```

```python
def replace_with_median(column_names):
    for i in range (len(column_names)):
        median_value = copied_dataset[column_names[i]].median()
        copied_dataset[column_names[i]] = copied_dataset[column_names[i]].fillna(value = median_value)
    return

def replace_with_none(column_names):
    for i in range(len(column_names)):
        copied_dataset[column_names[i]] = copied_dataset[column_names[i]].fillna(value = 'None')
    return

def replace_with_zero(column_names):
    for i in range(len(column_names)):
        copied_dataset[column_names[i]] = copied_dataset[column_names[i]].fillna(value = 0)
    return
```

```python
column_names = ['HDL cholesterol','d-dimer','Heart rate','HBB','Diuresis','Platelets']
replace_with_median(column_names)
```

```python
column_names = ['Mode_transport','Occupation']
replace_with_none(column_names)
```

```python
column_names = ['Children','FT/month']
replace_with_zero(column_names)
```

## scaling and  encoding the features checkpoint_2

```python
unscaled_dataset = copied_dataset.copy()
```

```python
location = unscaled_dataset['Region']

dummies = pd.get_dummies(location)

unscaled_dataset = pd.concat([unscaled_dataset,dummies],axis = 1)

unscaled_dataset = unscaled_dataset.drop(['Region'],axis =1)
```

```python
unscaled_dataset['Married'] = unscaled_dataset['Married'].map({'YES':1,'NO':0})
```

## Occupation feature

```python
occupation_dummies = pd.get_dummies(unscaled_dataset['Occupation'])

occupation_dummies.columns.values

columns_occ = [ 'Cleaner', 'Clerk', 'Driver',  'Legal',
       'Manufacturing','Business', 'None', 'Researcher','Farmer','Sales']

occupation_dummies = occupation_dummies[columns_occ]
```

```python
occ_group1 = occupation_dummies.iloc[:,0:5].max(axis=1)

occ_group2 = occupation_dummies.iloc[:,6:].max(axis=1)

unscaled_dataset = pd.concat([unscaled_dataset,occ_group1,occ_group2],axis=1)

unscaled_dataset = unscaled_dataset.drop(['Occupation'],axis = 1)
```
